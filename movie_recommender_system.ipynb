```python
def convert_column(col):
    lst = []
    for i in ast.literal_eval(col):
        lst.append(i['name'])
    return lst
```

```python
import ast
```

```python
merged_df['genres'] = merged_df['genres'].apply(convert_column)
```

```python
merged_df['keywords'] = merged_df['keywords'].apply(convert_column)
```

```python
merged_df.head()
```

```python
merged_df['cast'].head(1)
```

```python
def convert_cast(col):
    lst = []
    count = 0
    for i in ast.literal_eval(col):
        if count!=3:
            lst.append(i['name'])
            count+=1
        else:
            break
            
    return lst
```

```python
merged_df['cast'].apply(convert_cast)
```

```python
merged_df['cast'] = merged_df['cast'].apply(convert_cast)
```

```python
merged_df.head(1)
```

```python
def convert_crew(col):
    lst = []
    for i in ast.literal_eval(col):
        if i['job'] == 'Director':
            lst.append(i['name'])
            break
    return lst
```

```python
merged_df['crew'].apply(convert_crew)
```

```python
merged_df['crew'] = merged_df['crew'].apply(convert_crew)
```

```python
merged_df.head(1)
```

```python
merged_df['overview'] = merged_df['overview'].apply(lambda x:x.split())
```

```python
merged_df.head(1)
```

```python
merged_df.dtypes
```

```python
merged_df.head(1)
```

```python
merged_df['genres'] = merged_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
merged_df['overview'] = merged_df['overview'].apply(lambda x:[i.replace(" ","") for i in x])
merged_df['cast'] = merged_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
merged_df['crew'] = merged_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
merged_df.head(1)
```

```python
merged_df.drop('tags',axis=1,inplace=True)
```

```python
merged_df['tags'] = merged_df['overview'] + merged_df['genres'] + merged_df['keywords'] + merged_df['cast'] + merged_df['crew']
```

```python
merged_df.head(1)
```

```python
final_movies_df = merged_df[['id','title','tags']]
final_movies_df.head(1)
```

```python
final_movies_df['tags'][0]
```

```python
final_movies_df['tags'] = final_movies_df['tags'].apply(lambda x:" ".join(x))
```

```python
final_movies_df['tags'] = final_movies_df['tags'].apply(lambda x:x.lower())
```

```python
final_movies_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(final_movies_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# we found an isuue with words having same meaning but in different tense so we will use stemming to solbe that
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# example of stemming
ps.stem('dancing')
```

```python
ps.stem('running')
```

```python
def stem_func(text):
    lst = []
    for i in text.split():
       lst.append(ps.stem(i))
    
    return " ".join(lst) 
```

```python
final_movies_df['tags'].apply(stem_func)
```

```python
final_movies_df['tags'] = final_movies_df['tags'].apply(stem_func)
```

```python
vectors = cv.fit_transform(final_movies_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# gives the distance of each movie from all other movies basically similarity between the movies

similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
def recommed_movies(movie_name):
    
    movie_index = final_movies_df[final_movies_df['title'] == movie_name].index[0]
    distances = similarity[movie_index]
    recommended_movies = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in recommended_movies:
        print(final_movies_df.iloc[i[0]].title)

```
